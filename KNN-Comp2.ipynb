```python
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
X = iris.data; y = iris.target
scores = np.empty([20, 25])
k_range = range(1, 26)

for i in range(20):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=4)
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        scores[i,k-1] = metrics.accuracy_score(y_test, y_pred)
```

```python
#plot first test
plt.plot(k_range, scores[0,:],'r')
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy (First Test)')
plt.grid(1)
```

```python
#plot Last test
plt.plot(k_range, scores[-1,:],'r')
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy (Last Test)')
plt.grid(1)
```

```python
#plot first test
plt.plot(k_range, scores.mean(axis=0))
plt.xlabel('Value of K for KNN mean')
plt.ylabel('Testing Accuracy')
plt.grid(1)
```
